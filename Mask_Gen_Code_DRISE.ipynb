```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import torch
from torch import nn
from torch.nn import functional as F
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_masks(N, h, w, H, W, p):
    masks = torch.zeros(N, H, W)
    Ch = np.floor(H/h)
    print(Ch)
    Cw = np.floor(W/w)
    print(Cw)
    x_up = int(Ch*(h+1))
    print(x_up)
    y_up = int(Cw*(w+1))
    print(y_up)
    
    for i in range(N):
        mask = torch.zeros(h, w)
        mask = mask.bernoulli_(p)
        # Bilinear Interpolation
        mask = F.interpolate(mask.unsqueeze(0).unsqueeze(0), size=(x_up, y_up), mode='bilinear', align_corners=False)
        mask = mask.squeeze()
        # Random Cropping with uniformly random offsets
        x_offset = np.random.randint(0, x_up-H)
        y_offset = np.random.randint(0, y_up-W)
        mask = mask[x_offset:x_offset+H, y_offset:y_offset+W]
        masks[i] = mask
        # Saving the mask
        cv2.imwrite('Masks2/mask'+str(i)+'.png', mask.numpy()*255)
        print(f"Mask {i} generated and saved")
    return masks
```

```python
# Sample generation
N = 5000
h = 20
w = 20
H = 375
W = 500
p = 0.5
masks = generate_masks(N, h, w, H, W, p)

# fig, axs = plt.subplots(1, 5, figsize=(20, 6))
# for i in range(N):
#     axs[i].imshow(masks[i], cmap='gray')
#     axs[i].axis('off')
# plt.show()
```
